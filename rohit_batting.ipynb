```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load datasets (Replace 'file_path' with actual file paths in your Colab environment)
odi_df = pd.read_csv("ODI.csv")
t20_df = pd.read_csv("T20.csv")
test_df = pd.read_csv("TEST.csv")

# Display first few rows of each dataset
print("ODI Stats:")
display(odi_df.head())
print("T20 Stats:")
display(t20_df.head())
print("Test Stats:")
display(test_df.head())
```

```python
# Check for missing values
print("Missing Values in ODI:")
print(odi_df.isnull().sum())
print("Missing Values in T20:")
print(t20_df.isnull().sum())
print("Missing Values in Test:")
print(test_df.isnull().sum())
```

```python
# Basic Statistics
print("\nODI Statistics")
print(odi_df.describe())
print("\nT20 Statistics")
print(t20_df.describe())
print("\nTest Statistics")
print(test_df.describe())
```

```python
# Visualizing Runs per Year
plt.figure(figsize=(12,6))
sns.lineplot(data=odi_df, x='Year', y='Runs', marker='o', label='ODI')
sns.lineplot(data=t20_df, x='Year', y='Runs', marker='s', label='T20')
sns.lineplot(data=test_df, x='Year', y='Runs', marker='^', label='Test')
plt.title('Rohit Sharma - Runs Scored Per Year')
plt.xlabel('Year')
plt.ylabel('Runs')
plt.legend()
plt.grid()
plt.show()
```

```python
print(odi_df.columns)
print(t20_df.columns)
print(test_df.columns)
```

```python
odi_df.columns = odi_df.columns.str.lower()
t20_df.columns = t20_df.columns.str.lower()
test_df.columns = test_df.columns.str.lower()
```

```python
print("ODI SR count:", odi_df['sr'].count())
print("T20 SR count:", t20_df['sr'].count())
print("Test SR count:", test_df['sr'].count())
```

```python
t20_df['sr'] = t20_df['sr'].fillna(t20_df['sr'].median())
test_df['sr'] = test_df['sr'].fillna(test_df['sr'].median())
```

```python
odi_df['sr'] = pd.to_numeric(odi_df['sr'], errors='coerce')
t20_df['sr'] = pd.to_numeric(t20_df['sr'], errors='coerce')
test_df['sr'] = pd.to_numeric(test_df['sr'], errors='coerce')
```

```python
# Strike Rate Comparison
plt.figure(figsize=(12,6))
sns.boxplot(data=[odi_df['sr'], t20_df['sr'], test_df['sr']], palette='Set2')
plt.xticks([0,1,2], ['ODI', 'T20', 'Test'])
plt.title('Strike Rate Distribution Across Formats')
plt.ylabel('Strike Rate')
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(10,6))
# Exclude non-numeric columns before calculating correlation
numeric_odi_df = odi_df.select_dtypes(include=np.number)
sns.heatmap(numeric_odi_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("ODI Stats Correlation Heatmap")
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(10,6))
# Exclude non-numeric columns before calculating correlation
numeric_odi_df = t20_df.select_dtypes(include=np.number)
sns.heatmap(numeric_odi_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("ODI Stats Correlation Heatmap")
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(10,6))
# Exclude non-numeric columns before calculating correlation
numeric_odi_df = test_df.select_dtypes(include=np.number)
sns.heatmap(numeric_odi_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("ODI Stats Correlation Heatmap")
plt.show()
```

```python
# Calculating Not Outs
odi_df['not_out'] = odi_df['innings'] - odi_df['outs']
t20_df['not_out'] = t20_df['innings'] - t20_df['outs']
test_df['not_out'] = test_df['innings'] - test_df['outs']
```

```python
# Visualizing Runs per Year
plt.figure(figsize=(12,6))
sns.lineplot(data=odi_df, x='year', y='runs', marker='o', label='ODI')
sns.lineplot(data=t20_df, x='year', y='runs', marker='s', label='T20')
sns.lineplot(data=test_df, x='year', y='runs', marker='^', label='Test')
plt.title('Rohit Sharma - Runs Scored Per Year')
plt.xlabel('Year')
plt.ylabel('Runs')
plt.legend()
plt.grid()
plt.show()
```

```python
# Not Out Analysis
oti_not_outs = odi_df.groupby('year')['not_out'].sum()
t20_not_outs = t20_df.groupby('year')['not_out'].sum()
test_not_outs = test_df.groupby('year')['not_out'].sum()

plt.figure(figsize=(12,6))
sns.lineplot(x=oti_not_outs.index, y=oti_not_outs.values, marker='o', label='ODI Not Outs')
sns.lineplot(x=t20_not_outs.index, y=t20_not_outs.values, marker='s', label='T20 Not Outs')
sns.lineplot(x=test_not_outs.index, y=test_not_outs.values, marker='^', label='Test Not Outs')
plt.title('Not Out Innings Per Year')
plt.xlabel('Year')
plt.ylabel('Not Out Innings')
plt.legend()
plt.grid()
plt.show()
```

```python
# Consistency Analysis
odi_df['runs_std'] = odi_df['runs'].rolling(window=3).std()
plt.figure(figsize=(12,6))
sns.lineplot(data=odi_df, x='year', y='runs_std', marker='o', label='ODI Consistency')
plt.title('Consistency Analysis: Standard Deviation of Runs (3-year window)')
plt.xlabel('Year')
plt.ylabel('Standard Deviation of Runs')
plt.legend()
plt.grid()
plt.show()
```

```python
# Distribution of 50s and 100s
fig, axes = plt.subplots(1, 3, figsize=(18,6))
sns.barplot(data=odi_df, x='year', y='100', ax=axes[0], color='b')
axes[0].set_title('ODI Centuries Per Year')
sns.barplot(data=t20_df, x='year', y='100', ax=axes[1], color='g')
axes[1].set_title('T20 Centuries Per Year')
sns.barplot(data=test_df, x='year', y='100', ax=axes[2], color='r')
axes[2].set_title('Test Centuries Per Year')
plt.tight_layout()
plt.show()
```

```python
# Runs vs Balls Faced
plt.figure(figsize=(12,6))
sns.regplot(data=odi_df, x='balls', y='runs', scatter_kws={'alpha':0.5})
plt.title('ODI: Runs vs Balls Faced')
plt.xlabel('Balls Faced')
plt.ylabel('Runs Scored')
plt.grid()
plt.show()
```

```python
# Boundary Analysis
plt.figure(figsize=(12,6))
sns.lineplot(data=odi_df, x='year', y='4s', marker='o', label='Fours')
sns.lineplot(data=odi_df, x='year', y='6s', marker='s', label='Sixes')
plt.title('Boundaries Per Year (ODI)')
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend()
plt.grid()
plt.show()
```

```python
# Rolling Average of Runs
odi_df['rolling_avg_runs'] = odi_df['runs'].rolling(window=3).mean()
plt.figure(figsize=(12,6))
sns.lineplot(data=odi_df, x='year', y='rolling_avg_runs', marker='o', label='ODI')
plt.title('Rolling Average Runs (3-year window)')
plt.xlabel('Year')
plt.ylabel('Runs')
plt.legend()
plt.grid()
plt.show()
```

```python
# Rolling Average of Runs
odi_df['rolling_avg_runs'] = t20_df['runs'].rolling(window=3).mean()
plt.figure(figsize=(12,6))
sns.lineplot(data=odi_df, x='year', y='rolling_avg_runs', marker='o', label='ODI')
plt.title('Rolling Average Runs (3-year window)')
plt.xlabel('Year')
plt.ylabel('Runs')
plt.legend()
plt.grid()
plt.show()
```

```python
# Rolling Average of Runs
odi_df['rolling_avg_runs'] = test_df['runs'].rolling(window=3).mean()
plt.figure(figsize=(12,6))
sns.lineplot(data=odi_df, x='year', y='rolling_avg_runs', marker='o', label='ODI')
plt.title('Rolling Average Runs (3-year window)')
plt.xlabel('Year')
plt.ylabel('Runs')
plt.legend()
plt.grid()
plt.show()
```

```python
# Strike Rate Trend Over Time
plt.figure(figsize=(12,6))
sns.lineplot(data=odi_df, x='year', y='sr', marker='o', label='ODI')
sns.lineplot(data=t20_df, x='year', y='sr', marker='s', label='T20')
sns.lineplot(data=test_df, x='year', y='sr', marker='^', label='Test')
plt.title('Strike Rate Trend Over Time')
plt.xlabel('Year')
plt.ylabel('Strike Rate')
plt.legend()
plt.grid()
plt.show()
```

```python
# Dot Ball Percentage vs Runs
plt.figure(figsize=(12,6))
sns.scatterplot(data=odi_df, x='dot %', y='runs', alpha=0.5)
plt.title('Dot Ball Percentage vs Runs (ODI)')
plt.xlabel('Dot Ball Percentage')
plt.ylabel('Runs')
plt.grid()
plt.show()
```
